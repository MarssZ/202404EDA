```python
# 超参数
COLS_Y = 'order_cnt'
COLS_ALL = ['amount','order_cnt', 'code_day' , 'code_night', 'high', 'low', 'encoded_wc_night', 'code_week', 'code_holidays','encoded_holidays', '305_price','encoded_305_price','encode_weight_305price','313_price' ,'encoded_313_price']

# 读取excel文件并获取指定列
import pandas as pd
df = pd.read_csv('Data\结果1_全站按日期分组_2023全年.csv', usecols=COLS_ALL, encoding='GBK')

# 计算相关系数
correlation_matrix = df.corr()

# 绘制热图
import seaborn as sns
import matplotlib.pyplot as plt
# 设置图像大小
plt.figure(figsize=(12, 9)) 
# 设置坐标轴标签字体大小
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```
