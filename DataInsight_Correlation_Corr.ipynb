```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", message="Glyph .* missing from current font.")

# 读取excel文件并获取指定列
df = pd.read_csv('Data\store_6922.csv', usecols=['amt', 'discount_amount', 'pay_amount', 'num', 'discount', 'discount_percent'], encoding='GBK')

# 将列的数值赋值给数组array_X
array_amt = df['amt'].values
```

```python
# 使用默认的pearson相关系数
correlation_matrix  = df.corr()
print(correlation_matrix )

#画出相关性的热力图
plt.rcParams['font.family'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
sns.heatmap(correlation_matrix, annot=True)
plt.title('Heat Map', fontsize=18)
```

```python
# Plot
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8), dpi= 80)
sns.pairplot(
    df, 
    kind="scatter",     # "scatter" 散点图、 "reg"（绘制带有回归线的散点图）、"resid"（绘制残差图）、"kde"（绘制核密度估计图）等。
    #hue="species",      # 按 名为 species 的列的类别着色。
    plot_kws=dict(
        s=50,               # 散点图点的大小
        edgecolor="white",  # 散点图边框颜色
        linewidth=2.5)      # 散点图边框宽度
    )
plt.show()
```
